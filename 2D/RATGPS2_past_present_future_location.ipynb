```python
import pickle
from statistics import mean
import matplotlib.pyplot as plt
```

```python
with open('Mean_error_RFR_hippocampus', 'rb') as f:
    MAE_RFR_63 = pickle.load(f)

with open('Mean_error_RFR_entorhinal_cortex', 'rb') as f:
    MAE_RFR_37 = pickle.load(f)
    
with open('Mean_error_RFR_both', 'rb') as f:
    MAE_RFR_both = pickle.load(f)
```

```python
len(MAE_RFR_63), len(MAE_RFR_37), len(MAE_RFR_both)
```

## Hippocampus

```python
MAE_RFR_63_past_mean = mean(MAE_RFR_63[0:300])
MAE_RFR_63_present = MAE_RFR_63[300]
MAE_RFR_63_future_mean = mean(MAE_RFR_63[301:601])
MAE_RFR_63_past_mean, MAE_RFR_63_present, MAE_RFR_63_future_mean
```

```python
timesteps = list(range(-300,301))

plt.rcParams["figure.figsize"] = (20, 8)
plt.scatter(timesteps, MAE_RFR_63, s = 3)
plt.hlines(y = MAE_RFR_63_past_mean, xmin = -300, xmax = -1, colors = "red")
plt.hlines(y = MAE_RFR_63_future_mean, xmin = 1, xmax = 300, colors = "red")
plt.title('Mean absolute error of Random Forest Regression')
plt.xlabel('Timesteps (200ms)')
plt.ylabel('MAE')
plt.show()
```

```python
MAE_RFR_63_past_mean = mean(MAE_RFR_63[250:300])
MAE_RFR_63_present = MAE_RFR_63[300]
MAE_RFR_63_future_mean = mean(MAE_RFR_63[301:351])
MAE_RFR_63_past_mean, MAE_RFR_63_present, MAE_RFR_63_future_mean
```
